```python
from sklearn.datasets import load_breast_cancer
data=load_breast_cancer()
data
```

```python
len(data.target)
```

```python
import pandas as pd
X=pd.DataFrame(data.data,columns=data['feature_names'])
X
```

```python
X['target']=data.target
```

```python
X.head(5)
```

```python

# In case you want labels instead of numbers.
X.replace(to_replace={'target': {0:'malignant'}}, inplace=True)
X.replace(to_replace={'target': {1:'benign'}}, inplace=True)
```

```python
X
```

```python
X.isna().any()
```

```python
Z=X.corr()

pd.set_option('display.max_rows',500)
Z
```

```python
Z.rename(columns={'mean radius':'mean_radius'})
```

```python
from scipy.stats import zscore
Y=X.drop('target',axis=1)
Y.apply(zscore)
```

```python
y=Y.apply(sum)

print(y.values)
print(y.index)
```
