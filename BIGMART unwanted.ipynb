```python
def weight_pred(cols):        # after lot of repetative task created the user difined function for filling missing value 
    Iteam_Weight=cols[0]       #mean value of groupby ITEM_TYPE
    ITFV=cols[1]
    ITHD=cols[2]
    ITHH=cols[3]
    Item_Type_Household=cols[4]
    Item_Type_Meat=cols[5]
    Item_Type_Others=cols[6]
    Item_Type_Seafood=cols[7]
    ITSF=cols[8]
    ITSD=cols[9]
    ITSF=cols[10]
    
    if pd.isnull(Iteam_Weight):
        if ITFV==1:
            return 13.22
    if pd.isnull(Iteam_Weight):
        if ITHD==1:
            return 11.40
    if pd.isnull(Iteam_Weight):
        if ITHH==1:
            return 13.14
    if pd.isnull(Iteam_Weight):
        if Item_Type_Household==1:
            return 13.38
    if pd.isnull(Iteam_Weight):
        if Item_Type_Meat==1:
            return 12.81
    if pd.isnull(Iteam_Weight):
        if Item_Type_Others==1:
            return 13.85
    if pd.isnull(Iteam_Weight):
        if Item_Type_Seafood==1:
            return 12.55
    if pd.isnull(Iteam_Weight):
        if ITSF==1:
            return 12.98
    if pd.isnull(Iteam_Weight):
        if ITSD==1:
            return 11.84
    if pd.isnull(Iteam_Weight):
        if ITSF==1:
            return 13.69
    else:
        return Iteam_Weight
```

```python
X['Item_Weight']=X[['ITFV','ITHD','ITHH','Item_Type_Household','Item_Type_Meat','Item_Type_Others','Item_Type_Seafood','ITSF','ITSD','ITSF']].apply(weight_pred,axis=1)
#applying fun to X df
```
